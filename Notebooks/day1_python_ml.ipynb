```python
df = pd.read_csv("Heart_Disease_Prediction.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(numeric_cols)
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print(categorical_cols)
```

```python
df_practice = df.copy()
numeric_cols = df_practice.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    df_practice.loc[df_practice.sample(frac=0.1).index, col] = np.nan
```

```python
df_practice.isnull().sum()
```

```python
categorical_cols = df_practice.select_dtypes(include=['object', 'category']).columns

for col in categorical_cols:
    df_practice.loc[df_practice.sample(frac=0.1).index, col] = np.nan
```

```python
df_practice.isnull().sum()
```

```python
for col in numeric_cols:
    df_practice[col].fillna(df_practice[col].mean(), inplace=True)
```

```python
df_practice.isnull().sum()
```

```python
for col in categorical_cols:
    df_practice[col].fillna(df_practice[col].mode()[0], inplace=True)
```

```python
df_practice.isnull().sum()
```
